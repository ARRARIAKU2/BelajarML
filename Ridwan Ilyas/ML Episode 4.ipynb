```python
#Tabel Logika AND 2 input
"""
x1  x2  y
1   1   1
1   0   0
0   1   0
0   0   0
"""
```

```python
# library ML
import keras
import numpy as np

# model
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[2])])

#model compile
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
# set data
xs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=int)
ys = np.array([0, 0, 0, 1], dtype=int)
```

```python
# tampilkan arsitektur awal
model.summary()
```

```python
weights = model.get_weights()
weights
```

```python
# traing
model.fit(xs, ys, epochs=500)
```

```python
data = [1, 1]
answer = model.predict(np.array([data]))
print(answer)
```

```python
weights = model.get_weights()
weights
```
